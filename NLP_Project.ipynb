```python
import nltk
import re
import math
import pandas as pd
from collections import defaultdict
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.probability import FreqDist

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
    
from nltk.corpus import stopwords
from collections import defaultdict
```

```python
corpus = []
pos_index = defaultdict(dict)
doc_freq = defaultdict(int)
freq = defaultdict(dict)
docword = defaultdict()
```

```python
companies=['Adobe','Amazon','Facebook','Flipkart','Google','GoldmanSachs','Microsoft','MorganStanley','Samsung','TCS']
topics=['General', 'DFS', 'Oops', 'Web', 'DBMS', 'String', 'Error', 'Java', 'LinkedList', 'BFS', 'Sorting', 'Stack', 'Matrix' , 'Python', 'C','Array','OS','Queue','Bitwise',
        'ComputerNetwork','Aptitude','Email','DP']

for com in companies:
  path = "/content/%s_IR.txt"%(com)
  print(path)

  sample = open(path, "r").read()
  corpus.append(sample)

  sample = sample.lower()      
  tokenized_words = word_tokenize(sample)
  stopped_words = stopwords.words('english')

  # removing punctuation and stop words
  punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~+='''
  tokens = [word for word in tokenized_words if word not in punc]
  tokens = [word for word in tokens if word not in stopped_words]
  tokens = [word for word in tokens if word.isalpha()]
 
  porter=PorterStemmer()

  freq[com] = FreqDist(word for word in tokens)
  docword[com] = len(tokens)
       
 
  print(com)
  print(tokens)

  num = 0
  for x in tokens:
    filename = "%s_IR" %(com)
    x = porter.stem(x)
    if filename not in pos_index[x].keys():   #x = word: [ filename : position]
        pos_index[x].update({filename:[num]})
    else:
        pos_index[x][filename].append(num)
    num+=1

```

```python
def docfreq():    #no of documents a particular word is in 
  global doc_freq
  for term in pos_index.keys():
    doc_freq[term] = len( pos_index[term].values() )

docfreq()

#print(doc_freq.keys())
#print(doc_freq.values())

def termfreq(docID, word):    # no of time a word is in the particular docID
  if word in pos_index.keys():
    if docID in pos_index[word]:
      raw = len(pos_index[word][docID])
      
      #print(docID + " termfreq " + str(raw))
      return 1 + math.log(raw,10)
    else:
      return 0.0

  else:
    return 0.0

#print(pos_index['web'])
#print(termfreq('Google_IR','web'))


def tf(query, word):    # term freq of word in a query 
  if word in query:
    raw = query.count(word)   # raw term frequency
    #print(word + " tf - " + str(raw))
    return 1+math.log(raw,10)   # impact 
  else:
    return 0.0

def idf(word,docID):   # inverse document frequency for word in particular docID
  #print("word is " + str(word))
  #print("docID is "+ str(docID))
  if (docID+"_IR") in pos_index[word].keys():  #(N/0) error
    #print(docID + " idf - "+ (math.log(docword[docID]/doc_freq[word],10)))
    #print("idf")
    return math.log(docword[docID]/doc_freq[word],10)   #log(N/df)
  else:
    return 0.0

def tf_idf(query,word,docID):  #td-idf
  if word in pos_index.keys():
    #print("tf-idf")
    return idf(word,docID)* tf(query,word) 
  else:
    return 0
```

```python
def inputProcessing(word):
    porter=PorterStemmer()
    word=word.casefold()
    word = porter.stem(word)
    return word

print(topics)
query = list(input().split(" "))
query1 = []
for word in query:
  if word in topics:
    print("workss")
    query1.append(inputProcessing(word))

print(query1)

result = {}

for i in companies:
  
  cosine = {}  #dictionary

  for word in query1:
    docID = str(i) + "_IR"
    cosine[word] = termfreq(docID, word)     #word : term freq in query
    res = sum(map(lambda j : j * j, cosine.values() ))**(1/2)   # under root( sum of all tf-weights )
    
    #print(res)
    if res != 0:
      cosine[word] = (cosine[word]/res)*tf_idf(query1, word,i)   #normalised 
    #  print(word, cosine[word])

  result[i] = sum(cosine.values())   #similarity score 
  
print(result)

from collections import Counter

d = Counter(result)
d.most_common()

flag = 0
for k, v in d.most_common(5):
  if v!= 0:
    print (str(k) + "  " + str(v))
      
  
```
